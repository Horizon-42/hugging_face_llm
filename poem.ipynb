# English Poem Creater

```python
# login in huggingface-cli
import os
from huggingface_hub import login
from dotenv import load_dotenv
load_dotenv()  # 默认读取项目根的 .env
hf_token = os.getenv("HF_TOKEN")
if hf_token:
    login(token=hf_token)
else:
    print("HF_TOKEN not found in environment variables.")
```

```python
from transformers import pipeline

# checkpoint = "meta-llama/Meta-Llama-3-8B" #limited access
checkpoint = "mistralai/Mistral-7B-v0.1"
generator = pipeline("text-generation", model=checkpoint)

# Ask the user for three words (comma separated)
words = input("Enter three words separated by commas (e.g. ocean, moon, silence): ")
w = [s.strip() for s in words.split(",") if s.strip()][:3]
while len(w) < 3:
    w.append(input(f"Enter word {len(w)+1}: ").strip())

prompt = (
    f"Write a short poem (4-8 lines) in natural English that smoothly includes the "
    f"following three words at least once each: {w[0]!r}, {w[1]!r}, {w[2]!r}. "
    "Do not list the words or explain them — just the poem."
)


results = generator(prompt, max_length=50, num_return_sequences=1)

print("=" * 40)
print("the generated poem is:\n")
print(results[0]['generated_text'])
```
